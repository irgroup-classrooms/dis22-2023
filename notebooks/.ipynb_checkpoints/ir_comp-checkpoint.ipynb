````python
from querygpt.model import Query, Persona
````

````python
# Persönlicher Key muss eingesetzt werden und vorm pushen wieder entfernen
import openai
openai.api_key = ""
````

````python
import json
import pandas as pd
core_nist = json.load(open("../QueryGPT/core_nist.json"))
df = pd.DataFrame(columns= ['index','queries','topic', 'persona'])

for name in Persona.list_register():
    person = Persona.from_register(name)
    q = Query()
    q.select_querie_type("QUERY_EXAMPLE")
    q.select_persona(person)
    q.try_query_mappings()
    i = 0
    for topic in core_nist[1:6]:
        i +=1
        query = q.generate_query(topic=topic["title"], description=topic["desc"], narrative=topic["narr"], n_queries=10)
        completion = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "user", "content": query}
            ],
            temperature=0.2
            )
        answer = eval(completion.choices[0].message.content)
        df_temp = pd.DataFrame.from_dict(answer['queries'], orient='index', columns= ['queries']).reset_index()
        df_temp['topic'] = i
        df_temp['persona'] = name
        df = pd.concat([df, df_temp])
        print(df)
        
        
````

````python
df.to_csv('5topic.csv')
````

## IR Analysis of Personas

````python
import pandas as pd
persona_queries = pd.read_csv("5topic.csv")

persona_queries = persona_queries.drop("Unnamed: 0", axis = 1).rename(columns = {"index": "qno", "persona": "user", "topic": "qid", "queries": "query"})
persona_queries = persona_queries[["qno", "user", "qid", "query"]]
uqv = persona_queries
uqv
````

````python
import os
# needs to be adjusted to own system
os.environ["JAVA_HOME"] = "C:\Program Files\Java\jdk-20" #"/usr/lib/jvm/java-11-openjdk-amd64"

import pyterrier as pt
if not pt.started():
    pt.init()
    

from ranx import Qrels, Run, evaluate, compare, fuse, optimize_fusion
    
dataset = pt.get_dataset('irds:nyt/trec-core-2017')

#needs to be adjusted to own system
index_ref = pt.IndexRef.of('./../../index/nyt') 

qrels = Qrels.from_ir_datasets("nyt/trec-core-2017")

bm25 = pt.BatchRetrieve(index_ref, wmodel='BM25')
baseline = bm25(dataset.get_topics('title'))
pt.io.write_results(baseline, 'tmp', format='trec', append=False)
baseline_ranx = Run.from_file('tmp', kind='trec')
baseline_ranx.name = 'BM25'

runs = []

for i in range(1, 11):

  queries = uqv[uqv['qno']==i][['qid', 'query']]

  pt.io.write_results(bm25(queries), 'tmp', format='trec', append=False)
  runs.append(Run.from_file('tmp', kind='trec'))

fuse_method = 'rrf'

run_rrf = fuse(runs=runs, method=fuse_method)
run_rrf.name = 'RRF (BM25)'

report = compare(
    qrels=Qrels.from_ir_datasets("nyt/trec-core-2017"),
    runs=[baseline_ranx, run_rrf],
    metrics=["ndcg", "map", "precision@10", "bpref"],
    max_p=0.05  
)

report
````
